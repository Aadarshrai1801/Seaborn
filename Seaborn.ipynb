### Data Visualization with Seaborn

```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn
tips = sns.load_dataset("tips")
tips
```

```python
# Create a Scatter Plot
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot of Total bill vs tips")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```

```python
# Line Plot
sns.lineplot(x="size",y="total_bill",data=tips)
plt.title("Scatter Plot of Size vs Total Bill")
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Categorical Plots

# Bar Plot
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Bar Plot of Day vs Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Box Plot
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Box Plot of Day vs Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Violin Plot
sns.violinplot(x="day",y="total_bill",data=tips)
plt.title("Violin Plot of Day vs Total Bill")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Histogram
sns.histplot(tips["total_bill"],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips["total_bill"],fill=True)
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
# Heat Map
corr = tips[["total_bill","tip","size"]].corr()
corr
```

```python
# Heat Ma
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
